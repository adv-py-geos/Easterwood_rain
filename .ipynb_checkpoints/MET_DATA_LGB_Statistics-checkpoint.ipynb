```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Column List
name_list = ["Station ID","Date [UTC]","Temp [F]","DP [F]","RH [%]","W Dir [Deg]","W Spd [Kts]","Alt [inHg]","1Hr-Prcp [mm]",
             "Vis [mi]","SKC1","SKC2","SKC3","Cld Hgt1 [Ft]","Cld Hgt2 [Ft]","Cld Hgt3 [Ft]","Prs Wx"]

# Data Type List
dtype_list = {"Station ID":"str","UTC":"str","Temp [F]":"float64","DP [F]":"float64","RH [%]":"float64","W Dir [Deg]":"float64",
              "W Spd [Kts]":"float64","Alt [inHg]":"float64","1Hr-Prcp [mm]":"float64","Vis [mi]":"float64","SKC1":"str",
              "SKC2":"str","SKC3":"str","Cld Hgt1 [Ft]":"float64","Cld Hgt2 [Ft]":"float64","Cld Hgt3 [Ft]":"float64",
              "Prs Wx" : "str"}

# Date Column
parse_date = ["Date [UTC]"]

# Import TxT Data
CLL = pd.read_csv('Data/CLL.txt', sep='\t', header=0, names = name_list, dtype = dtype_list, parse_dates = parse_date, index_col = 1).resample('1d').mean()
DFW = pd.read_csv('Data/DFW.txt', sep='\t', header=0, names = name_list, dtype = dtype_list, parse_dates = parse_date, index_col = 1).resample('1d').mean()
AUS = pd.read_csv('Data/AUS.txt', sep='\t', header=0, names = name_list, dtype = dtype_list, parse_dates = parse_date, index_col = 1).resample('1d').mean()
IAH = pd.read_csv('Data/IAH.txt', sep='\t', header=0, names = name_list, dtype = dtype_list, parse_dates = parse_date, index_col = 1).resample('1d').mean()
```

```python
CLL.head()
```

```python
#Calculate Mean, Median, and Standard deviation

CLL_mean=CLL.mean().to_frame().T
CLL_med=CLL.median().to_frame().T
CLL_std=CLL.std().to_frame().T

DFW_mean=DFW.mean().to_frame().T
DFW_med=DFW.median().to_frame().T
DFW_std=DFW.std().to_frame().T

AUS_mean=AUS.mean().to_frame().T
AUS_med=AUS.median().to_frame().T
AUS_std=AUS.std().to_frame().T

IAH_mean=IAH.mean().to_frame().T
IAH_med=IAH.median().to_frame().T
IAH_std=IAH.std().to_frame().T
```

```python
CLL_STAT=CLL_mean
CLL_STAT=CLL_STAT.append([CLL_med.loc[0],CLL_std.loc[0]], ignore_index=True)
CLL_STAT.insert(0,"STAT",['MEAN','MEDIAN' ,'STD'])

DFW_STAT=DFW_mean
DFW_STAT=DFW_STAT.append([DFW_med.loc[0],DFW_std.loc[0]], ignore_index=True)
DFW_STAT.insert(0,"STAT",['MEAN','MEDIAN' ,'STD'])

AUS_STAT=AUS_mean
AUS_STAT=AUS_STAT.append([AUS_med.loc[0],AUS_std.loc[0]], ignore_index=True)
AUS_STAT.insert(0,"STAT",['MEAN','MEDIAN' ,'STD'])

IAH_STAT=IAH_mean
IAH_STAT=IAH_STAT.append([IAH_med.loc[0],IAH_std.loc[0]], ignore_index=True)
IAH_STAT.insert(0,"STAT",['MEAN','MEDIAN' ,'STD'])
```

```python
#Naming dataframes
CLL.name='CLL'
DFW.name='DFW'
AUS.name='AUS'
IAH.name='IAH'

def gen_hist(a,nbins):
    fig, axs = plt.subplots(2, 2,figsize = (10, 10))
    fig.suptitle('Histograms of '+a.name+' Data', fontsize=15)
    nbins=nbins
   
    #(1)Temp
    axs[0,0].hist(a["Temp [F]"],bins=nbins)
    axs[0,0].set_title("Temp [F]")
    axs[0,0].set_xlabel('Temp [F]')
    axs[0,0].axvline(a["Temp [F]"].mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = axs[0,0].set_ylim()
    axs[0,0].text(a["Temp [F]"].mean()*0.65, max_ylim*0.9, 'Mean: {:.2f}'.format(a["Temp [F]"].mean()))
    #axs[0,0].grid()
    
    #(2)Alt [inHg]
    axs[0,1].hist(a["Alt [inHg]"],bins=nbins)
    axs[0,1].set_title("Alt [inHg]")
    axs[0,1].set_xlabel('Alt [inHg]')
    axs[0,1].axvline(a["Alt [inHg]"].mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = axs[0,1].set_ylim()
    axs[0,1].text(a["Alt [inHg]"].mean()*1.002, max_ylim*0.9, 'Mean: {:.2f}'.format(a["Alt [inHg]"].mean()))
    #axs[0,1].grid()
    
    #(3)1Hr-Prcp [mm]
    axs[1,0].hist(a["1Hr-Prcp [mm]"],bins=nbins)
    axs[1,0].set_title("1Hr-Prcp [mm]")
    axs[1,0].set_xlabel('1Hr-Prcp [mm]')
    axs[1,0].axvline(a["1Hr-Prcp [mm]"].mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = axs[1,0].set_ylim()
    axs[1,0].text(a["1Hr-Prcp [mm]"].mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(a["1Hr-Prcp [mm]"].mean()))
    #axs[1,0].grid()

    #(4)"W Spd [Kts]
    axs[1,1].hist(a["W Spd [Kts]"],bins=nbins)
    axs[1,1].set_title("W Spd [Kts]")
    axs[1,1].set_xlabel('W Spd [Kts]')
    axs[1,1].axvline(a["W Spd [Kts]"].mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = axs[1,1].set_ylim()
    axs[1,1].text(a["W Spd [Kts]"].mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(a["W Spd [Kts]"].mean()))
    #axs[1,1].grid()
```

```python
gen_hist(DFW,50)
```

```python
gen_hist(IAH,50)
```

```python
gen_hist(AUS,50)
```
